```python
from textblob import TextBlob
```

```python
data.head()
```

```python
TextBlob("I've been following you from the start of your").sentiment.polarity
```

```python
polarity=[]
for i in data['comment_text']:
    sentiment_score=TextBlob(i).sentiment.polarity
    try:
        polarity.append(sentiment_score)
    except:
        polarity.append(0)
```

```python
polarity[1:10]
```

```python
data['polarity']=polarity
```

```python
data.head()
```

```python
len(polarity)
```

```python
import emoji
```

```python
data['comment_text'].head()
```

```python
len(data['comment_text'])
```

```python
emoji_list=[]
for comment in data['comment_text'].dropna():
  for char in comment:
    if char in emoji.EMOJI_DATA:
       emoji_list.append(char)
```

```python
len(emoji_list)
```

```python
emoji_list[1:10]
```

```python
from collections import Counter
```

```python
Counter(emoji_list).most_common(10)
```

```python
Counter(emoji_list).most_common(10)[0]
```

```python
emoji_sticker=[]
emoji_count=[]
for i in Counter(emoji_list).most_common(10):
    emoji_sticker.append(i[0])
    emoji_count.append(i[1])
```

```python
emoji_sticker
```

```python
emoji_count
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
trace=go.Bar(x=emoji_sticker,y=emoji_count)
```
